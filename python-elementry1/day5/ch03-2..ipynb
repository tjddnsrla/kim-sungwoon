```python
treeHit = 0
while treeHit < 10:
    treeHit = treeHit + 1
    print("나무를 %d번 찍었습니다." % treeHit)
    if treeHit == 10:
        print("나무 넘어갑니다.")
```

```python
# while 반복문 빠져나가기 방법
# 1. while문의 조건식이 False일 때
# 2. break문을 활용하여 while문 빠져나가기
coffee = 10
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee - 1
    print("남은 커피의 양은 %d개입니다." % coffee)

    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
```

```python
coffee = 10
while True:
    money = int(input("돈을 넣어 주세요: "))

    if money == 300:
        print("커피를 줍니다.")
        coffee = coffee - 1
    elif money > 300:
        print("거스름돈 %d를 주고 커피를 줍니다." % (money - 300))
        coffee = coffee - 1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print("남은 커피의 양은 %d개 입니다." % coffee)
    
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지 합니다.")
        break
```

```python
# while문의 맨 처음으로 돌아가기 -> continue문
a = 0
while a < 10:
    a = a + 1
    if a % 2 == 0: 
        continue
    print(a)
```

```python
# 문제1
# while문을 사용해서 1부터 1000까지 3의 배수의 합을 화면에 출력하세요.
# Hint: 3의 배수는 3으로 나누었을 때 나머지 값이 0인 것
a = 0
b = 0
while a < 1001:
    a += 1
    if a % 3 == 0:
        print("3의 배수", a)
        b += a

print("3의 배수의 합", b)
```

```python
# 문제2
# while문을 사용해서 1이상 100 이하의 짝수를 모두 출력해 보세요
# Hint: 짝수는 2으로 나누었을 때 나머지 값이 0인 것
# 2
# 4
# 6
a = 1
#while 1 <= a <= 100:
while a >= 1 and a <= 100:
    if a % 2 == 0:
        print(a)
    a += 1
```

```python
# 문제4
# while문과 if문을 활용하여, 100 이하의 자연수 중 8의 배수이지만 12의 배수는 아닌 것을 모두 출력하세요.
# 예를 들어서 16은 8의 배수이지만 12의 배수가 아니니가 조건에 부합합니다. 
# 하지만 48은 8의 배수이면 12의 배수이기도 해서 조건에 부합하지 않습니다.
# 출력
# 8
# 16
# 32
# .
# .
# 88

a = 1
while a < 101:
    if a % 8 == 0 and a % 12 != 0: print(a)
    a += 1
```

```python
# 문제5
# 10보다 작은 2 또는 3의 배수는 2,3,4,6,8,9이며 이들의 합은 32입니다.
# while문과 if문을 활용하여 1000보다 작은 자연수 중 2 또는 3의 배수의 합을 출력하는 프로그램을 만드세요
# 출력 : 333167
i = 1
sum = 0

while i < 1000:
    if i % 2 == 0 or i % 3 == 0:
        sum = sum + i
    i += 1

print(sum)
```

```python
# 문제6
# 정수 n의 약수는 n을 나누었을 때 나누어 떨어지는 수입니다. 
# 만약 정수 i가 정수 n의 약수라면 n을 i로 나누었을 때 나머지가 0이 나와야 하는 거죠.
# 정수 120의 약수를 모두 출력하고 총 몇 개의 약수가 있는지 출력하는 프로그램을 만드세요
# Hint:  120 % i == 0
i = 1
count = 0

while i < 121:
    if 120 % i == 0:
        print(i)    # 약수값
        count += 1
    i += 1   

print("120 약수의 총 개수는", count)
```

```python
# 문제7
# 이중 while문을 사용해서 구구단 프로그램을 만들어 보세요.
# 실행하면 아래와 같이 출력이 되어야 합니다.
# 1 * 1 = 1
# 1 * 2 = 2
# 1 * 3 = 3

# 2 * 1 = 2
# 2 * 2 = 4
# 2 * 3 = 6

# .
# .

# 9 * 1 = 9
# 9 * 2 = 18
# 9 * 3 = 27

a = 1
b = 1
while a < 10:
    print(a, "단", sep = '')        
    while b < 10:   # a = 2, b = 10 -> 1
        print("{} * {} = {}".format(a, b, a * b))
        b += 1
    a += 1      # a = 2, b = 10
    b = 1
```

```python
# 문제8
# 피보나치 수열문제
# 1, 1, 2, 3, 5, 8, 13, 21, 45... 와 같이 1번 항과 2번 항은 각각 1로 세팅이 되고 
# 3번 항부터는 바로 앞 두 항의 합으로 계산되는 수열이다. 피보나치 수열의 첫 50개 항을 차례대로 출력하는 프로그램을 작성

prev = 0
current = 1
i = 1

while i < 51:
    print(current)
    
    temp = prev
    prev = current
    current = current + temp

    i += 1
```
